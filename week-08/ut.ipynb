```python
ut_df=ut_df.rename(columns={'birthstate': 'birthplace'})
ut_df.sample(10
        )
```

```python
def replace_married(some_string):
   if some_string == 'm':
     return "married"
   if some_string == 's':
    return "single"
   elif some_string == 'w':
    return "widowed"
```

```python
ut_df['marrystatus']=ut_df['marrystatus'].apply(replace_married)
ut_df
```
